```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

## df = pd.read_csv('sample_data.csv')

# Sample time-based data with at least 10 data points
np.random.seed(42)
start = pd.Timestamp('2022-01-01')
end = pd.Timestamp('2022-01-20')  # Extend the date range
data1 = pd.date_range(start, end, freq='D')
data2 = pd.date_range(start + pd.Timedelta(days=5), end + pd.Timedelta(days=5), freq='D')  # Offset the second dataset
numeric_data1 = (data1 - start).days
numeric_data2 = (data2 - start).days

# Create a whisker plot
plt.figure(figsize=(10, 6))  # Increase figure size

# First whisker plot
plt.boxplot([numeric_data1, numeric_data2], vert=False)

# Add labels and title
plt.xlabel('Days from Start Date', labelpad=15)

plt.ylabel('Date', rotation=0, labelpad=10, verticalalignment='top')  # Rotate y-axis label by 90 degrees, move to the left, and a little bit up
plt.gca().yaxis.set_label_coords(-0.1,0.5)

plt.title('Whisker Plot Example for Time-Based Data')

# Remove right and top spines
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

plt.savefig('whisker_plot.png', dpi=300, bbox_inches='tight')


plt.show()
```
